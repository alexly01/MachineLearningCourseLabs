```python
import numpy as np
import matplotlib.pyplot as plt

#variables

#alpha steplengths
alpha1 = 1
alpha2 = 0.1
alpha3 = 0.01

#initial point
w0 = 2

#max steps
K = 1000

#func g
def g(w):
    return 1/50 * (w**4 + w**2 + 10 * w)

#func deriv of g
def deriv_g(w):
    return 1/50 * (4 * w**3 + 2 * w + 10)

#func gradient descent
def gradient_descent(g, deriv_g, alpha, K, w0):
    w_list = [w0]
    g_list = [g(w0)]
    
    for k in range(K):
        w = w_list[-1] - alpha * deriv_g(w_list[-1])
        w_list.append(w)
        g_list.append(g(w))

    return w_list, g_list

#run gradient descent using different steplengths
w_list1, g_list1 = gradient_descent(g, deriv_g, alpha1, K, w0)
w_list2, g_list2 = gradient_descent(g, deriv_g, alpha2, K, w0)
w_list3, g_list3 = gradient_descent(g, deriv_g, alpha3, K, w0)

#plot with  = 1
plt.figure()
plt.plot(g_list1)
plt.title(" = 1")
plt.xlabel("Iterations")
plt.ylabel("g(w)")
plt.show()

#plot with  = 0.1
plt.figure()
plt.plot(g_list2)
plt.title(" = 0.1")
plt.xlabel("Iterations")
plt.ylabel("g(w)")
plt.show()

#plot with  = 0.01
plt.figure()
plt.plot(g_list3)
plt.title(" = 0.01")
plt.xlabel("Iterations")
plt.ylabel("g(w)")
plt.show()

# plot all on the same figure
plt.figure()
plt.plot(g_list1, label=" = 1", color = "blue")
plt.plot(g_list2, label=" = 0.1", color = "red")
plt.plot(g_list3, label=" = 0.01", color = "green")
plt.title("Gradient Descent with Different  Values")
plt.xlabel("Iterations")
plt.ylabel("g(w)")
plt.legend()
plt.show()
```

Question: Which step length value works best for this particular function and initial points? Describe your answer.

The step length that works best for this particular function would most likely be alpha equaling 0.1. The reason I chose this one was because although alpha equaling 1 shows the fastest convergence/learning rate of the three, it could lead to some instability within the data meaning that sometimes it could jump over or skip certain regions of the data that should be taken a closer look upon. Also, with alpha equaling 1, it could overshoot the minimum or go past the optimal solution leading to faulty error within our data. As for alpha equaling 0.01, it takes too long before it reaches the minimum. As you can see as we're reaching our 1000 iteration, it still hasn't made its way and stabalized at the minimum. Alpha equaling 0.1's step lenghts works the best for this function since it does converge fast which is similar to the first, but it also has a smooth and stable convergence. It's not just a downward slope right to the minimum like alpha equaling 1.
