```python
#C) Linear Regression Model

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

# read in our excel file
data = pd.read_excel('Dry_Bean_Dataset.xlsx')

features = data.iloc[1:, :-1] 
labels = data.iloc[1:, -1]

# need to convert our labels to numerical format so can be read
label_encoder = LabelEncoder()
labels_float = label_encoder.fit_transform(labels)

# set up training and testing 
train_x, test_x, train_y, test_y = train_test_split(features, labels_float, test_size=0.2, shuffle=True)

scaler = StandardScaler()
# scale train_x
train_x_nor = scaler.fit_transform(train_x)
# scale test_x using train_x scaling
test_x_nor = scaler.transform(test_x)

# create linear regression model
lin_regr_model = LinearRegression()

# train our model onto the training data
lin_regr_model.fit(train_x_nor, train_y)

# predict our outputed labels from our model for both our training and testing set
train_predictions = lin_regr_model.predict(train_x_nor)
test_predictions = lin_regr_model.predict(test_x_nor)

# predictions need to be rounded so they can be classified into labels
train_pred_classes = [round(pred) for pred in train_predictions]
test_pred_classes = [round(pred) for pred in test_predictions]

# calculate our final accuracy
train_acc = accuracy_score(train_y, train_pred_classes)
test_acc = accuracy_score(test_y, test_pred_classes)

print("Accuracy of Training:", train_acc)
print("Accuracy of Testing:", test_acc)
```

```python
#D) Attempt 1: ANN

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt

# load the data
data = pd.read_excel('Dry_Bean_Dataset.xlsx')

features = data.iloc[1:, :-1]
labels = data.iloc[1:, -1]

# convert labels to numerical format
label_encoder = LabelEncoder()
labels_float = label_encoder.fit_transform(labels)

# set up training and testing 
train_x, test_x, train_y, test_y = train_test_split(features, labels_float, test_size=0.2, shuffle=True)

scaler = StandardScaler()
# scale train_x
train_x_nor = scaler.fit_transform(train_x)
# scale test_x using train_x scaling
test_x_nor = scaler.transform(test_x)

# create ANN model with our inputs
ANN = MLPClassifier(hidden_layer_sizes=(4), solver='sgd', learning_rate_init=0.8, max_iter=200, early_stopping=True, random_state =1)

# train our model onto the training data
ANN.fit(train_x_nor, train_y)

# get training accuracy
training_acc = ANN.score(train_x_nor, train_y)

# get test accuracy
test_acc = ANN.score(test_x_nor, test_y)

print("Attempt 1:")
print(f"Training Accuracy:", training_acc)
print(f"Test Accuracy:", test_acc)

plt.figure(figsize=(12, 6))
plt.plot(ANN.validation_scores_)
plt.title('Iteration vs Accuracy')
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.show()
```

```python
#D) Attempt 3: ANN

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt

# Load the data
data = pd.read_excel('Dry_Bean_Dataset.xlsx')

features = data.iloc[1:, :-1]
labels = data.iloc[1:, -1]

# Convert labels to numerical format
label_encoder = LabelEncoder()
labels_float = label_encoder.fit_transform(labels)

# set up training and testing 
train_x, test_x, train_y, test_y = train_test_split(features, labels_float, test_size=0.2, shuffle=True)

scaler = StandardScaler()
# scale train_x
train_x_nor = scaler.fit_transform(train_x)
# scale test_x using train_x scaling
test_x_nor = scaler.transform(test_x)

# create ANN model with our inputs
ANN = MLPClassifier(hidden_layer_sizes=(1), solver='sgd', learning_rate_init=0.1, max_iter=100, early_stopping=True, random_state =1)

# train our model onto the training data
ANN.fit(train_x_nor, train_y)

# get training accuracy
training_acc = ANN.score(train_x_nor, train_y)

# get test accuracy
test_acc = ANN.score(test_x_nor, test_y)

print("Attempt 3:")
print(f"Training Accuracy:", training_acc)
print(f"Test Accuracy:", test_acc)

plt.figure(figsize=(12, 6))
plt.plot(ANN.validation_scores_)
plt.title('Iteration vs Accuracy')
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.show()
```
